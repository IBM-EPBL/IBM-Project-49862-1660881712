# Outlier Removal using IQR

```python
q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
IQR = q3-q1
upper_limit = q3 + 1.5*IQR
lower_limit = q1 - 1.5*IQR
```

```python
df1=df[df.Age<upper_limit]
```

```python
sns.boxplot(df1.Age)
```

# Outlier Replacement: Median

```python
q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
IQR = q3-q1
upper_limit = q3 + 1.5*IQR
lower_limit = q1 - 1.5*IQR
```

```python
upper_limit
```

```python
df['Age'].median()
```

```python
df['Age']=np.where(df['Age']>upper_limit,df['Age'].median(),df['Age'])
```

```python
df.shape
```

```python
sns.boxplot(df.Age)
```

# Outlier Removal using Z-score

```python
from scipy import stats
```

```python
Age_zscore=stats.zscore(df.Age)
Age_zscore
```

```python
df_z=df[np.abs(Age_zscore)<=3]
df_z
```

```python
sns.boxplot(df_z.Age)
```
